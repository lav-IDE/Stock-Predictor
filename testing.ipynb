```python
import yfinance as yf

TICKER = "HDFCBANK"
ticker = yf.Ticker(TICKER+".NS")
news = ticker.news
print(news[0]['content'])
```

```python
import pandas as pd

rows = []
for i in news:
    rows.append({
        "date": pd.to_datetime(i["content"]["pubDate"]),
        "title": i["content"]["title"]
    })

yf_db = pd.DataFrame(rows, columns=["date", "title"])
yf_db["ticker"] = [TICKER for _ in range(len(yf_db))]

yf_db
```

```python
headline=yf_db.loc[5]
```

```python
from work.prediction_pipeline import StockMovementPredictor

predictor = StockMovementPredictor()
predictor.predict_batch(
    df=yf_db,
    headline_col='title'
)
```
